```python
data = pd.read_csv("charts_pop.csv")
data = data.iloc[:,1:]

data.head()
```

```python
lyrics_250 = []
counter_250 = 0
for song, artist in zip(data.song[:250], data.artist[:250]):
    lyrics = genius.get_lyrics(song, artist)
    lyrics_250.append(lyrics)
    counter_250 += 1
    print(counter_250)
```

```python
lyrics_250_df = pd.DataFrame()
lyrics_250_df['song'] = data.song[:250]
lyrics_250_df['aritst'] = data.artist[:250]
lyrics_250_df['lyrics'] = lyrics_250
```
